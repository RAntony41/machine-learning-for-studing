# Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
num_samples = 40

x_1 = np.random.multivariate_normal([2, 2], [[1, .75], [.75, 1]], num_samples)
y_1 = np.zeros(num_samples)

x_2 = np.random.multivariate_normal([-1, -1], [[1, .75], [.75, 1]], num_samples)
y_2 = np.ones(num_samples)

x = np.vstack((x_1, x_2))
y = np.hstack((y_1, y_2))

xplot = x.T
```

```python
plt.scatter(xplot[0],xplot[1], c='red')
#plt.plot()
plt.show()
```

```python
e = 2.71
def logistic_regression(x,y,epoch,learning):
    x_sample, x_feature = x.shape
    m = len(y)
    w = np.zeros(x_feature)
    b = 0.0
    
    for i in range(epoch):
        n = - np.dot(x,w)
        h = 1 / (1 + (e)**n)
        #print(i,h)
        dw = (1/m)*np.dot(x.T, h - y)
        db = (1/m)*np.sum(h - y)
        
        w -= learning * dw
        b -= learning * db
    
    return w,b
```

```python
w,b = logistic_regression(x,y,600,0.001)
print("w: ",w)
print("b: ",b)
```
