# Task List - Linear Regression Model with 1 Feature
0. Prepare your data - simulate straight line.
1. Compute the hypothesis
2. Find the Loss: MSE Error
3. Find the loss gradient for all thetas (model terms)
4. Update the thetas

```python
# Step 0 - Prepare your data - simulate straight line.

X = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
```

```python
# Step 1 - Compute the hypothesis


m = 2
c = -1

# h = [m * x + c for x in X]  # hypothesis

def computeHypothesis(X, m, c):
  h=[]
  for x in X:
    h.append(m*x+c)
  return h

computeHypothesis(X, m, c)
```

```python
# Step 2 - Compute the loss

def computeLoss(X,y,h):
  total_loss = 0
  for i in range( len(X) ):
    loss = ( float(y[i]) - h[i]) * ( float(y[i]) - h[i])
    total_loss = total_loss + loss

  mse = total_loss / ( 2*len(X) )
  return mse

computeLoss(X,y,h)
```

```python
# Step 3 - Find the gradient

# dm = derivative of gradient m = derivative of Loss with respect to theta1
# dc = derivative of gradient c = derivative of Loss with respect to theta0

def findGradient(X, y, h):
  dm_sum = 0
  dc_sum = 0
  for i in range( len(X) ):
    dm = (h[i]-y[i])*X[i]
    dm_sum = dm_sum + dm

    dc = (h[i]-y[i])
    dc_sum = dc_sum + dc

  dm_final = dm_sum / len(X)
  dc_final = dc_sum / len(X)

  return dm_final, dc_final

dm, dc = findGradient(X, y, h)
```

```python
# Step 4 - Update the thetas

# alpha is the learning rate, this is how fast you want to go down the gradient descent slope

def updateGradient( c, m, dc_final, dm_final ):

  alpha = 0.1

  m = m - alpha * dm_final
  c = c - alpha * dc_final

  return [m, c]

m, c = updateGradient( -1, 2, 3, 5 )
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plotResult(X,y,m,c):
  # Plotting the data points
  plt.scatter(X, y, color='blue', label='Data Points')

  # Plotting the line of best fit
  predicted_y = [m * x + c for x in X]
  plt.plot(X, predicted_y, color='red', label='Line of Best Fit')

  plt.xlabel('X')
  plt.ylabel('y')
  plt.title('Linear Regression with Gradient Descent')
  plt.legend()
  plt.show()
```

```python

X = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
m = 2
c = -1

h = computeHypothesis(X, m, c)
loss = computeLoss(X,y,h)

for i in range( 10 ):
  h = computeHypothesis(X, m, c)
  # print('h', h)
  # print('y', y)
  loss = computeLoss(X,y,h)
  dm, dc = findGradient(X, y, h)
  m, c = updateGradient( c, m, dc, dm )

  if (i%1 == 0):
    print('i', i)
    print('loss', loss)
    plotResult(X,y,m,c)
    print()
```
