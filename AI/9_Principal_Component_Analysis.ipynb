```python
# Standardization
def stand(x):
    return (x - x.mean()) / x.std()
```

```python
newdf = stand(df)
```

```python
# Covariance Matrix
def coMat(x):
    return np.cov(x.T)
```

```python
cov = coMat(df)
```

```python
cov
```

```python
# Evalue,Evector
def eigenV(x):
    Evalue,Evector = np.linalg.eig(x)
    #Evector = Evector.T
    return Evalue,Evector
```

```python
Evalue,Evector=eigenV(cov)
```

```python
def getmatrix(Evalue,Evector,n_com):
    sortV = np.argsort(Evalue)[::-1]
    Evalue = Evalue[sortV]
    Evector = Evector[sortV]
    
    return Evector[:n_com]
    
```

```python
n_com = 3
mat = getmatrix(Evalue,Evector,n_com)
```

# In class format

```python
class PCA():
    
    def __init__(self,n_com):
        self.n_com = n_com
        
    def fit(self,df):
        stan = self._stand(df)
        comM = np.cov(stan.T)
        Evalue,Evector = self._eigenV(comM)
        mat = self._getmatrix(Evalue,Evector,self.n_com)
        
        return np.dot(df,mat.T)

    def _stand(self,df):
        return (df - df.mean()) / df.std()
    
    def _eigenV(self,x):
        Evalue,Evector = np.linalg.eig(x)
        return Evalue,Evector
    
    def _getmatrix(self,Evalue,Evector,n_com):
        sortV = np.argsort(Evalue)[::-1]
        Evalue = Evalue[sortV]
        Evector = Evector[sortV]
    
        return Evector[:n_com] 
```

```python
model = PCA(6)
```

```python
pd.DataFrame(model.fit(df))
```
