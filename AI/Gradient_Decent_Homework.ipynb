```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv("/content/housing.csv",header=None, delimiter=r"\s+", names=column_names)
df = data
```

```python
df
```

```python
def normalize_dataframe(df):
    return (df - df.min()) / (df.max() - df.min())
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.3, random_state=42)

print("Train", train.shape)
print("Test:", test.shape)
```

```python
train
```

```python
train_y = train['MEDV']
train_x = train.drop(columns=['MEDV'])
```

```python
train_x = normalize_dataframe(train_x)
```

```python
test_y = test['MEDV']
test_x = test.drop(columns=['MEDV'])
```

```python
test_x = normalize_dataframe(test_x)
```

```python
def Gradient_decent2(x, y, learning_rate,epoch):
  x_sample, x_feature = x.shape
  x = np.array(x)
  y = np.array(y)

  # starting weight and bias
  weights = np.zeros(x_feature)
  bias = 0.0

  for i in range(epoch):
    hypo = np.dot(x,weights) + bias

    dw = 2*(1/(x_sample)) * np.dot(x.T,(hypo - y))
    db = 2*(1/x_sample) * np.sum(hypo - y)

    weights -= learning_rate * dw
    bias -= learning_rate * db

    '''
    print(i,hypo[0])
    print("dw", dw)
    print("db",db)
    print("weight", weights.tolist())
    print("bias",bias)
    '''

  print("weight", weights.tolist())
  print("bias",bias)
  return weights.tolist(),bias
```

```python
weight,bias = Gradient_decent2(train_x, train_y, 0.001,8000)
```

```python
def predict(x_test,y_test,weights,bias):
  y_pred = np.dot(x_test, np.array(weights) )+ bias
  #y_pred = [str(num)[:5] for num in y_pred]
  #y_pred = [float(num) for num in y_pred]
  mse = sum((y_test - y_pred)**2)/len(y_test)

  return mse
```

```python
print("Evaluation:")
predict(test_x,test_y,weight,bias)
```
